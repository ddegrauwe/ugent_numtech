```python
# load some libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
exercise='forward'
# read the output data
filename='osceq_'+exercise+'/output.dat'
data=np.loadtxt(filename,skiprows=1)
# select columns
t=data[:,0]
y_exact=data[:,1]
y_numerical=data[:,2]

# open a figure
fig, ax = plt.subplots(figsize=(12,8))
# plot exact and numerical result
plt.plot(t,y_exact,label='exact')
plt.plot(t,y_numerical,label=exercise)
# add legend
plt.legend();
```
